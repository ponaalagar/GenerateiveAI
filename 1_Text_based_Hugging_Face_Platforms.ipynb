```python
# ===========================================
# STEP 1: Install Hugging Face Transformers
# ===========================================
#google/vaultgemma-1b
!pip install transformers --quiet

# Import the pipeline API
from transformers import pipeline

# ===========================================
# 1. Sentiment Analysis
# ===========================================
print("\n=== Sentiment Analysis ===")
# Load a pre-trained sentiment analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis")

# Example texts to analyze
texts = [
    "I absolutely loved the movie! The acting was brilliant.",
    "The product was okay, nothing special.",
    "This service is terrible and I want a refund."
]

# Analyze each text for sentiment
for text in texts:
    result = sentiment_analyzer(text)[0]  # Get first (and only) result
    print(f"Text: {text}\nSentiment: {result['label']} (Score: {result['score']:.2f})\n")






```

```python
# 2. Text Generation

print("\n=== Text Generation ===")
# Load a small GPT-2 model for text generation
text_generator = pipeline("text-generation", model="gpt2-xl")

# Provide a starting text (prompt)
prompt = "Artificial Intelligence is transforming the world by"

# Generate continuation of the text
generated = text_generator(prompt, max_length=50, num_return_sequences=1)
print(f"Prompt: {prompt}\nGenerated: {generated[0]['generated_text']}\n")
```

```python
# 3. Text Summarization

print("\n=== Text Summarization ===")
# Load a summarization model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Provide a long text for summarization
long_text = """
Artificial Intelligence (AI) is rapidly transforming the way we live and work.
From healthcare and finance to entertainment and education, AI is helping to automate tasks,
provide insights, and improve efficiency. As AI technologies continue to evolve,
their impact on society will only grow, making it crucial for individuals and organizations
to understand and adapt to this transformation.
"""

# Generate a summary of the text
summary = summarizer(long_text, max_length=60, min_length=30, do_sample=False)
print(f"Original Text: {long_text}\n\nSummary: {summary[0]['summary_text']}\n")
```

```python

# 4. Named Entity Recognition (NER)

print("\n=== Named Entity Recognition (NER) ===")
# Load NER pipeline with grouped entities
ner = pipeline("ner", grouped_entities=True)

# Provide a sentence for entity extraction
sentence = "Apple was founded by Steve Jobs and Steve Wozniak in Cupertino, California in 1976."

# Perform NER on the sentence
entities = ner(sentence)
print(f"Text: {sentence}\nEntities:")
for entity in entities:
    print(f"- {entity['entity_group']}: {entity['word']} (Score: {entity['score']:.2f})")
print()
```

```python

# 5. Question Answering (Q&A)

print("\n=== Question Answering (Q&A) ===")
# Load Q&A pipeline with a distilled BERT model
qa = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

# Provide a context paragraph for Q&A
context = """
The Hugging Face Transformers library provides thousands of pre-trained models
to perform tasks on texts such as classification, information extraction, question answering, summarization, translation, and more.
"""

# Ask a question about the context
question = "What does the Hugging Face Transformers library provide?"

# Get the answer
answer = qa(question=question, context=context)
print(f"Context: {context}\nQuestion: {question}\nAnswer: {answer['answer']}")
```
